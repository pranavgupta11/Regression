```python

def get_rss(y_pred, y):  
    rss = np.sum((y_pred - y) ** 2)
    return rss
#TSS is the total sum of squares, sum of y_outcome - mean_y
#RSS is the residual sum of squares, sum of y_outcome - predicted_y
#R² = (TSS - RSS) / TSS
def get_r2(y_pred, y):  
    #rss = get_rss(y_pred, y)
    #tss = get_rss(y, np.mean(y))
    rss = np.sum((y_pred - y) ** 2)
    tss = np.sum((y - np.mean(y))** 2)
    r2=(tss-rss)/tss
    return r2

#linear regression - least square method
def estimate_coef(x, y):

	n = np.size(x)
	m_x = np.mean(x)
	m_y = np.mean(y)

	SS_xy = np.sum(y*x) - n*m_y*m_x
	SS_xx = np.sum(x*x) - n*m_x*m_x

	b_1 = SS_xy / SS_xx #slope
	b_0 = m_y - b_1*m_x #intercept

	return (b_0, b_1)

def plot_regression_line(x, y, b):
	# plotting the actual points as scatter plot
	plt.scatter(x, y, color = "m",marker = "o", s = 30)

	y_pred = b[0] + b[1]*x

	# plotting the regression line
	plt.plot(x, y_pred, color = "g")

	# putting labels
	plt.xlabel('x')
	plt.ylabel('y')

	plt.show()
 
def linear_regression(x,y):
    b = estimate_coef(x, y)
    print("Estimated coefficients:\nintercept = {} \nslope = {}".format(b[0], b[1]))
    plot_regression_line(x, y, b)
    return b

def get_regression_predictions(input_feature, intercept, slope):
    # calculate the predicted values:
    predicted_values = intercept + (slope * input_feature)
    return predicted_values

def linear_reg(feature):
    y,x= data["price"],data[feature]
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
    b=linear_regression(x_train,y_train)
    y_pred=get_regression_predictions(x_test,b[0],b[1])
    rss=get_rss(y_pred, y_test)
    rss = "{:e}".format(rss)
    print(f"RSS : {rss}\n")
    r2=get_r2(y_pred,y_test)
    print(f"R2 : {r2}\n")

#linear regression gradient descent 
def Linear_GradientDescent(x, y, b1, b0, learning_rate, epochs):
    cost_list = [0] * epochs
    
    m=len(y)
    for epoch in range(epochs):
        z =x*b1 + b0
        loss = z - y
        
        weight_gradient = x.T.dot(loss) / m
        bias_gradient = np.sum(loss) / m
        
        b1 = b1 - learning_rate*weight_gradient
        b0 = b0 - learning_rate*bias_gradient
  
        cost = (np.sum(((x*b1 + b0) - y) ** 2)) / (2*m)
        cost_list[epoch] = cost
        
        if (epoch%100==0):
            print(f"Cost at epoch {epoch} is : ","{:.2e}".format(cost))
        
    plt.plot(cost_list)
    plt.show()
    return b0,b1

def linear_reg_grad(feature):
    y,x= data["price"],data[feature]
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
    b=Linear_GradientDescent(x_train,y_train,0,0,1e-8,300)
    plot_regression_line(x_train,y_train,b)
    y_pred=get_regression_predictions(x_test,b[0],b[1])
    rss=get_rss(y_pred, y_test)
    rss = "{:e}".format(rss)
    print(f"RSS : {rss}\n")
    r2=get_r2(y_pred,y_test)
    print(f"R2 : {r2}\n")



#linear regression sklearn
def linear_reg_sklearn(feature):
    y,x= data["price"],data[feature]
    x=x.to_numpy()
    y=y.to_numpy()
    x=x.reshape(-1,1)
    y=y.reshape(-1,1)
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
    reg = LinearRegression().fit(x_train,y_train)
    b=[]
    slope=reg.coef_    
    intercept=reg.intercept_
    b.append(intercept)
    b.append(slope)
    y_pred=reg.predict(x_test)
    print(f"slope : {slope}")
    print(f"intercept : {intercept}")
    plot_regression_line(x_train,y_train,b)
    rss=get_rss(y_pred, y_test)
    rss = "{:e}".format(rss)
    print(f"RSS : {rss}\n")
    r2=r2_score(y_test,y_pred)
    print(f"R2 : {r2}\n")

#Linear Regression - Analysis of scores with different features
def cumulative_reg():
    rss_list=[]
    col_list=[]
    r2_list=[]
    i=0
    for col in data.columns:
        if(i>2):
            y,x= data["price"],data[col]
            x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
            b=estimate_coef(x_train,y_train)
            y_pred=get_regression_predictions(x_test,b[0],b[1])
            rss=get_rss(y_pred, y_test)
            r2=get_r2(y_pred,y_test)
            rss_list.append(rss)
            col_list.append(col)
            r2_list.append(r2)  
        i=i+1
    
    get_bar_chart(rss_list,col_list,'RSS')
    get_bar_chart(r2_list,col_list,'R2')

#multivariate linear regression with gradient descent optimization

def Multivariate_CostFunction(x,y,w,b,m):
    cost = (np.sum(((x.dot(w) + b) - y) ** 2)) / (2*m)
    return cost
"""
w : theta (1 to num of features)
b : theta naught 

"""
def Multivariate_GradientDescent(x, y, w, b, learning_rate, epochs):
    cost_list = [0] * epochs
    
    m=len(y)
    for epoch in range(epochs):
        z =np.dot(x,w) + b
        loss = z - y
        
        weight_gradient = x.T.dot(loss) / m
        bias_gradient = np.sum(loss) / m
        
        w = w - learning_rate*weight_gradient
        b = b - learning_rate*bias_gradient
  
        cost = Multivariate_CostFunction(x, y, w, b,m)
        cost_list[epoch] = cost
        
        if (epoch%100==0):
            print(f"Cost at epoch {epoch} is : ","{:.2e}".format(cost))
        
    #plt.plot(np.arrange(epochs),cost_list)
    plt.plot(cost_list)
    plt.show()
    
    return w, b

def multivariate_prediction(X, w, b):
    return X.dot(w) + b

def mul_reg(feature_list):
    x=data[feature_list]
    y=data["price"]
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)

    x=x.to_numpy()
    y=y.to_numpy()

    w, b= Multivariate_GradientDescent(x_train, y_train, np.zeros(x_train.shape[1]), 0, 1e-9,epochs=2000)

    y_pred = multivariate_prediction(x_test, w, b)
    RSS=get_rss(y_pred, y_test)
    RSS= "{:e}".format(RSS)
    print(f"RSS : {RSS}\n")
    r2=get_r2(y_pred,y_test)
    print(f"R2 : {r2}\n")


def get_line():
    print("\n")
    for i in range(0,100):
        print('*',end="")
    print("\n")



#multivariate regression - sklearn
def mul_reg_sklearn(feature_list):
    x=data[feature_list]
    y=data["price"]
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)

    x=x.to_numpy()
    y=y.to_numpy()
    x=x.reshape(-1,1)

    reg = LinearRegression().fit(x_train,y_train)
    w=[]
    w=reg.coef_    
    b=reg.intercept_
    y_pred=reg.predict(x_test)
    rss=get_rss(y_pred, y_test)
    rss = "{:e}".format(rss)
    print(f"RSS : {rss}\n")
    r2=r2_score(y_test,y_pred)
    print(f"\nR2 : {r2}")
    

def get_bar_chart(rss_list,col_list,score):

    fig = plt.figure(figsize = (20, 7))
    # creating the bar plot
    plt.bar(col_list, rss_list)
    plt.xlabel("Features")
    plt.ylabel(f"{score} Score")
    plt.show()



def main():
    #main
    ch=1
    while (ch!=7):
        get_line()
        print("""
                        Welcome to Seattle House Price Prediction Simulator

        1. Linear Regression - Least Squares Method
        2. Linear Regression - Gradient Descent Optimization
        3. Linear Regression - sklearn
        4. Linear Regression - Analysis of scores with different features
        5. Multivariate Linear Regression - Gradient Descent Optimization    
        6. Multivariate Linear Regression - sklearn
        7. Exit \n
        """)
        ch=int(input("Enter your choice : "))

        if(ch==1):
            get_line()
            print("-----------LINEAR REGRESSION - Least Squares Method---------")
            #linear regression
            print("features : ")
            i=0
            for col in data.columns:
                if(i>2):
                    print(col,end="\t")
                i=i+1
                if(i%12==0):
                    print()
            f=input("\nEnter feature name : ")
            linear_reg(f)
        elif(ch==2):
            get_line()
            print("-----------LINEAR REGRESSION - Gradient Descent---------")
            #linear regression
            print("features : ")
            i=0
            for col in data.columns:
                if(i>2):
                    print(col,end="\t")
                i=i+1
                if(i%12==0):
                    print()
            f=input("\nEnter feature name : ")
            linear_reg_grad(f)
        elif(ch==3):
            get_line()
            print("\n-----------LINEAR REGRESSION - sklearn---------")
            print("features : ")
            i=0
            for col in data.columns:
                if(i>2):
                    print(col,end="\t")
                i=i+1
                if(i%12==0):
                    print()
            f=input("\nEnter feature name : ")
            linear_reg_sklearn(f)
        elif(ch==4):
            get_line()
            print("\n-----------LINEAR REGRESSION - Analysis of scores with different features---------")
            cumulative_reg()

        elif(ch==5):
            get_line()
            print("-----------MULTIVARIATE REGRESSION - Gradient Descent---------")
            #multivariate regression
            print("features : ")
            i=0
            for col in data.columns:
                if(i>2):
                    print(col,end="\t")
                i=i+1
                if(i%12==0):
                    print()
            f_lst = []
            n = int(input("\nEnter number of features : "))
            print("\nEnter feature names : ")
            for i in range(0, n):
                ele = input()
                f_lst.append(ele)
            
            mul_reg(f_lst)       
        
        elif(ch==6):
            get_line()
            print("\n-----------MULTIVARIATE REGRESSION - sklearn---------")
            print("features : ")
            i=0
            for col in data.columns:
                if(i>2):
                    print(col,end="\t")
                i=i+1
                if(i%12==0):
                    print()
            n = int(input("\nEnter number of features : "))
            f_lst = []
            print("\nEnter feature names : ")
            for i in range(0, n):
                ele = input()
                f_lst.append(ele)
            
            mul_reg_sklearn(f_lst)

        elif(ch==7):
            get_line()
            print("\nThank You.")
        else:
            get_line()
            print("\nInvalid Input. Try Again.\n")

if __name__=="__main__":
    main()
```
